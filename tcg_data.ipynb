```python
import re
import pandas as pd

data = pd.read_excel("merch-pre-order.xlsx")
purchased = data["Kindly select the merch you are willing to avail. Product mock ups are for reference only and may slightly vary.: Products"]

packs = {"Starter":0, "Booster":0}

# Initial stock qty
init_starter = 30
init_booster = 40

# Current stock qty
curr_starter = 2
curr_booster = 2

for item in purchased:
    products = item.splitlines()[:-1]

    for product in products:
        if "OVERLOAD: The Official DCS Trading Card Game" in product:
            parameters = re.search(r'\((.*?)\)', product)
            parameters_str = parameters.group(1)
            data = parameters_str.split(", ")[1:]
            quantity = int(data[0].strip("Quantity: "))
            type_pack = data[1].strip("Pack: ")
            packs[type_pack] += quantity

# Print the data
print("Initial Stock")
print(f"Starter: {init_starter}")
print(f"Booster: {init_booster}")
print("")

print("Quantity Bought")
for key, val in packs.items():
    print(f"{key}: {val}")
print("")

print("Current Stock")
print(f"Starter: {curr_starter}")
print(f"Booster: {curr_booster}")
print("")

print("Deduction")
print(f"Starter: {init_starter - packs["Starter"] - curr_starter}")
print(f"Booster: {init_booster - packs["Booster"] - curr_booster}")
```
